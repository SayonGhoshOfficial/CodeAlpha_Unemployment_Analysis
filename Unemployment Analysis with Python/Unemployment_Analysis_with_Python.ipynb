## 🧪 Step 1: Import Required Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## 📂 Step 2: Load the Dataset
Upload the file `Unemployment_Rate_upto_11_2020.csv` in the left-hand Files panel.

```python
import os
print(os.listdir())

# Make sure you have uploaded the Unemployment_Rate_upto_11_2020.csv file
# to your Colab environment before running this cell.
df = pd.read_csv('Unemployment_Rate_upto_11_2020.csv')
df.head()
```

## 🔍 Step 3: Explore the Dataset

```python
df.info()
df.describe()
df.columns
```

## 🧹 Step 4: Clean Column Names

```python
df.columns = ['Region', ' Date', ' Frequency', ' Estimated Unemployment Rate (%)',
              ' Estimated Employed', ' Estimated Labour Participation Rate (%)',
              'Region.1', 'longitude', 'latitude']
df.head()
```

## 🧼 Step 5: Check for Null or Missing Data

```python
df.isnull().sum()
```

## 📊 Step 6: Visualize Average Unemployment Rate by Region

```python
plt.figure(figsize=(12,6))
sns.barplot(x='Region', y=' Estimated Unemployment Rate (%)', data=df, errorbar=None)

plt.xticks(rotation=90)
plt.title('Average Unemployment Rate by Region')
plt.ylabel('Unemployment Rate (%)')
plt.tight_layout()
plt.show()
```

## 📈 Step 7: Visualize Unemployment Trend Over Time

```python
df[' Date'] = pd.to_datetime(df[' Date'], dayfirst=True)
plt.figure(figsize=(14,6))
sns.lineplot(data=df, x=' Date', y=' Estimated Unemployment Rate (%)', hue='Region', errorbar=None)

plt.title('Unemployment Rate Trend Across Regions')
plt.ylabel('Unemployment Rate (%)')
plt.xlabel('Date')
plt.tight_layout()
plt.show()
```
